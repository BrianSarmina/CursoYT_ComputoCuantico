```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
```

```python
# Definir el operador lineal (matriz 2x2)
A = np.array([[2, 1],
              [1, 2]])

# Calcular la traza y los autovalores
traza = np.trace(A)
autovalores, autovectores = np.linalg.eig(A)

print(f"Matriz A:\n{A}")
print(f"Traza de A: {traza}")
print(f"Autovalores de A: {autovalores}")
print(f"Autovectores de A:\n{autovectores}")

# Crear un círculo unitario
theta = np.linspace(0, 2*np.pi, 200)
circle = np.array([np.cos(theta), np.sin(theta)])  # Coordenadas del círculo

# Aplicar la transformación lineal al círculo
transformed_circle = A @ circle

# Configurar la gráfica
fig, ax = plt.subplots(figsize=(8, 8))

# Dibujar el círculo original
ax.plot(circle[0], circle[1], 'b', label='Círculo unitario')

# Dibujar el círculo transformado
ax.plot(transformed_circle[0], transformed_circle[1], 'r', label='Transformación de A')

# Dibujar los autovectores
origin = np.array([[0, 0], [0, 0]])  # Origen para los vectores
ax.quiver(*origin, autovectores[0, :], autovectores[1, :], color=['g','m'], scale=3, label='Autovectores')

# Configurar aspecto de la gráfica
ax.set_aspect('equal')
ax.grid(True)
ax.legend()
ax.set_title('Interpretación geométrica de la traza de un operador')
```
