```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
# Definir la matriz A
A = np.array([[2, 1],
              [1, 2]])

# Calcular eigenvalores y eigenvectores
eigenvalores, eigenvectores = np.linalg.eig(A)

print("Eigenvalores:")
print(eigenvalores)
print("\nEigenvectores:")
print(eigenvectores)

# Crear un conjunto de vectores aleatorios
np.random.seed(0)  # Para reproducibilidad
vectores = np.random.randn(5, 2)

# Aplicar la matriz A a los vectores
vectores_transformados = vectores @ A.T

# Graficar los vectores originales y transformados
plt.figure(figsize=(8, 8))
ax = plt.gca()

# Dibujar los eigenvectores
for i in range(len(eigenvalores)):
    eigenvector = eigenvectores[:, i] * eigenvalores[i]
    ax.arrow(0, 0, eigenvector[0], eigenvector[1],
             head_width=0.1, head_length=0.2, fc='red', ec='red', label=f'Eigenvector {i+1}')
    plt.text(eigenvector[0]*1.1, eigenvector[1]*1.1, f'Eigenvector {i+1}', color='red')

# Dibujar los vectores originales en azul
for v in vectores:
    ax.arrow(0, 0, v[0], v[1], head_width=0.05, head_length=0.1, fc='blue', ec='blue', alpha=0.5)

# Dibujar los vectores transformados en verde
for vt in vectores_transformados:
    ax.arrow(0, 0, vt[0], vt[1], head_width=0.05, head_length=0.1, fc='green', ec='green', alpha=0.5)


custom_handles = [
    Line2D([0], [0], color='red', lw=2, label='Eigenvectores'),
    Line2D([0], [0], color='blue', lw=2, label='Vectores originales'),
    Line2D([0], [0], color='green', lw=2, label='Vectores transformados')
]

# Configurar el gráfico
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Transformación de vectores por la matriz A')
plt.grid(True)
plt.legend(handles=custom_handles)
plt.axis('equal')
```
