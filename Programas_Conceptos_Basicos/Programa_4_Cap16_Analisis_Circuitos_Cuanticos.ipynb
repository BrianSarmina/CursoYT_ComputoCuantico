```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
```

```python
# Configurar el simulador
simulator = AerSimulator()

# Función para ejecutar y visualizar un circuito
def execute_and_visualize(qc, shots=1024):
    """Compila, ejecuta y visualiza los resultados de un circuito cuántico."""
    compiled_circuit = transpile(qc, simulator)
    result = simulator.run(compiled_circuit, shots=shots).result()
    counts = result.get_counts()
    return counts
```

```python
# 1. Circuito cuántico que aplica la compuerta X de Pauli
print("1. Circuito cuántico con compuerta X:")
qc_x = QuantumCircuit(1, 1)
qc_x.x(0)  # Aplicar compuerta X
qc_x.measure(0, 0)  # Medir el qubit
print(qc_x.draw())
counts_x = execute_and_visualize(qc_x)
plot_histogram(counts_x)
```

```python
# 2. Circuito cuántico que aplica la compuerta Z de Pauli
print("2. Circuito cuántico con compuerta Z:")
qc_z = QuantumCircuit(1, 1)
qc_z.x(0)  # Aplicar X
qc_z.z(0)  # Aplicar compuerta Z
qc_z.measure(0, 0)  # Medir el qubit
print(qc_z.draw())
counts_z = execute_and_visualize(qc_z)
plot_histogram(counts_z)
```

```python
# 3. Circuito cuántico que aplica la compuerta H
print("3. Circuito cuántico con compuerta H:")
qc_h = QuantumCircuit(1, 1)
qc_h.h(0)  # Aplicar compuerta H
qc_h.measure(0, 0)  # Medir el qubit
print(qc_h.draw())
counts_h = execute_and_visualize(qc_h)
plot_histogram(counts_h)
```

```python
# 4. Circuito cuántico que aplica H, Z y H
print("4. Circuito cuántico con H, Z y H:")
qc_hzh = QuantumCircuit(1, 1)
qc_hzh.h(0)  # Aplicar compuerta H
qc_hzh.z(0)  # Aplicar compuerta Z
qc_hzh.h(0)  # Aplicar compuerta H de nuevo
qc_hzh.measure(0, 0)  # Medir el qubit
print(qc_hzh.draw())
counts_hzh = execute_and_visualize(qc_hzh)
plot_histogram(counts_hzh)
```
